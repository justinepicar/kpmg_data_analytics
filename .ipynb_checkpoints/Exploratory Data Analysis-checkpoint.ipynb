```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import xlrd
```

```python
xlsx = pd.ExcelFile('KPMG_clean_data.xlsx')
df = pd.read_excel(xlsx, 'ExistingCustomerList')
new_customers = pd.read_excel(xlsx, 'NewCustomerList')
```

# CATEGORICAL AND NUMERICAL DISTRIBUTIONS

```python
cat_features = [ 'gender', 'job_title', 'job_industry_category', 'wealth_segment',
               'deceased_indicator','owns_car', 'tenure', 'postcode', 'state', 
               'online_order', 'order_status', 'brand', 'product_line', 'product_class', 
               'product_size', 'age']
num_features = ['past_3_years_bike_related_purchases', 'property_valuation', 'list_price', 'standard_cost']
```

```python
df[num_features].describe().T
```

```python
df[num_features].agg(['skew', 'kurtosis']).T
```

```python
# change postcode to a string so we can use a countplot to view the distributions
df.postcode = df.postcode.astype('str')
```

```python
for i in num_features:
    fig, axes = plt.subplots(1, 1, figsize=(10,10))
    sns.histplot(data = df, x = df[i])
    plt.title(f'{i} Distributions')
    plt.show()
```

```python
for i in cat_features:
    fig, axes = plt.subplots(1, 1, figsize=(10,10))
    counts_index = df[i].value_counts().index
    if len(counts_index) > 10:
        top10 = df[i].value_counts().head(10)
        sns.barplot(y=top10.index, x=top10.values)
        plt.title(f'Top 10 {i} Distributions')
    else:
        sns.countplot(data = df, y = df[i], order=df[i].value_counts().index)
        plt.title(f'{i} Distributions')
    plt.show()

# change postcode back to an object
df.postcode = df.postcode.astype('object')
```

# TRANSACTION DATES BY FEATURES

```python
def transaction_date_lineplot(dt_feat, feat):
    
    plt_df = df.loc[:, [dt_feat, feat]]
    plt_df['qty'] = 1
    plt_df = plt_df.loc[:, [dt_feat, feat, 'qty']]
    plt_grouped = plt_df.groupby([dt_feat, feat]).count().reset_index()
    plt_wide = plt_grouped.pivot(dt_feat, feat, 'qty')
    
    plt.figure(figsize = (15,10))
    sns.lineplot(data=plt_wide)
    plt.title(f'{dt_feat} by {feat}')
    plt.show()
```

```python
transactions = df.transaction_date.value_counts()

plt.figure(figsize=(15,10))
sns.lineplot(x=transactions.index, y=transactions.values)
plt.title('Date of Transactions Jan 2017-Dec 2017')
plt.show()
```

```python
transaction_date_lineplot('transaction_date', 'brand')
transaction_date_lineplot('transaction_date', 'product_line')
transaction_date_lineplot('transaction_date', 'age')
transaction_date_lineplot('transaction_date', 'tenure')
transaction_date_lineplot('transaction_date', 'state')
```
